### Renaming personality trait columns for easier readibility 

```python
data.rename(columns=lambda col: col.split('_', 1)[1] if '_' in col else col, inplace=True)
data.rename(columns={'sex': 'Cat_sex'}, inplace=True)
```

### Removing null values

```python
data[data['Cat_sex'].isnull() == True]
```

Only the `Cat_sex` column has null values, and there are only 23 instances so they will be removed. 

```python
data = data.dropna()
```

```python
data
```

### Checking values for each column
#### Expected values:
 - The 52 personality traits should be integers ranging from 1-7
 - Country should be New Zealand or Australia 
 - Cat_sex should be male or female 

```python
for column in data.columns:
    print(data[column].unique())
```

```python
# Inspecting the datapoints where the cat sex is 'unsure'

data[data['Cat_sex'] == 'Unsure']
```

There are only 15 instances of the cat's sex being unsure, so those datapoints will be removed as well. 

```python
data = data[data['Cat_sex'] != 'Unsure']
```

```python
# Validating cleaned dataset

data
```

The final cleaned dataset has 2764 rows and 54 columns, 38 rows were removed during preprocessing. The personality trait columns were also renamed to remove the `Personality#_` prefix for easier readibility

```python
# Saving cleaned dataset to directory

current_dir = os.getcwd()
main_dir = os.path.abspath(os.path.join(current_dir, '..'))
data_dir = os.path.abspath(os.path.join(main_dir, 'data'))
data.to_csv(os.path.join(data_dir, 'data_preprocessed.csv'), index=False)
```
